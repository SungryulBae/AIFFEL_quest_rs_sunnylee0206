# GD04 Ablation Study — Gaussian Sigma

| 변수 | 조건 A | 조건 B (베이스라인) | 조건 C |
|---|---|---|---|
| Gaussian Sigma | 1 | **2** | 3 |

- 고정값: EPOCHS=5, BATCH=16, LR=7e-4, Model=SimpleBaseline(ResNet-50), SEED=42

## 2. Import & 시드 고정

```python
import random, os, math, io, json
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.models as tv_models
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from PIL import Image
from torch.utils.data import Dataset, DataLoader

SEED = 42

def set_seed(seed=SEED):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark     = False

def seed_worker(worker_id):
    worker_seed = (torch.initial_seed() + worker_id) % 2**32
    np.random.seed(worker_seed)
    random.seed(worker_seed)

def make_generator():
    g = torch.Generator()
    g.manual_seed(SEED)
    return g

set_seed()
print(f"SEED={SEED} 고정 완료")
print('슝=3')
```

## 3. Config

```python
PROJECT_PATH = os.path.join(os.getenv("HOME"), 'work/mpii')
IMAGE_PATH   = os.path.join(PROJECT_PATH, 'images')
MODEL_PATH   = os.path.join(PROJECT_PATH, 'models')
TRAIN_JSON   = os.path.join(PROJECT_PATH, 'mpii_human_pose_v1_u12_2', 'train.json')
VALID_JSON   = os.path.join(PROJECT_PATH, 'mpii_human_pose_v1_u12_2', 'validation.json')
os.makedirs(MODEL_PATH, exist_ok=True)

IMAGE_SHAPE  = (256, 256, 3)
HEATMAP_SIZE = (64, 64)
NUM_HEATMAP  = 16
EPOCHS       = 5
BATCH_SIZE   = 16
LR           = 7e-4
POS_WEIGHT   = 81

# ── 팀원 3 실험 조건 ──────────────────────────────────────────────────────────
SIGMA_CONDITIONS = {
    'SIGMA_A': 1,   # 조건 A
    'SIGMA_B': 2,   # 조건 B ← 베이스라인
    'SIGMA_C': 3,   # 조건 C
}

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device : {device}")
print(f"Epochs={EPOCHS} | Batch={BATCH_SIZE} | LR={LR}")
print(f"Sigma 조건: {SIGMA_CONDITIONS}")
print('슝=3')
```

## 4. Preprocessor (sigma 파라미터화)

```python
class Preprocessor:
    def __init__(self, image_shape=(256,256,3), heatmap_shape=(64,64,16),
                 is_train=False, gaussian_sigma=1):
        self.is_train       = is_train
        self.image_shape    = image_shape
        self.heatmap_shape  = heatmap_shape
        self.gaussian_sigma = gaussian_sigma  # ← 실험 포인트

    def __call__(self, example):
        features = self.parse_tfexample(example)
        image    = Image.open(io.BytesIO(features['image/encoded']))
        if self.is_train:
            margin = torch.empty(1).uniform_(0.1, 0.3).item()
            image, kp_x, kp_y = self.crop_roi(image, features, margin)
        else:
            image, kp_x, kp_y = self.crop_roi(image, features)
        image    = image.resize((self.image_shape[1], self.image_shape[0]))
        img_np   = np.array(image).astype(np.float32) / 127.5 - 1.0
        img_t    = torch.from_numpy(img_np).permute(2, 0, 1)
        heatmaps = self.make_heatmaps(features, kp_x, kp_y, self.heatmap_shape)
        return img_t, heatmaps

    def parse_tfexample(self, example):
        anno   = example['annotation']
        joints = anno['joints']
        jv     = anno.get('joints_vis', [1]*len(joints))
        return {
            'image/encoded'         : self.image_to_bytes(example['image']),
            'image/object/parts/x'  : [j[0] for j in joints],
            'image/object/parts/y'  : [j[1] for j in joints],
            'image/object/parts/v'  : jv,
            'image/object/center/x' : anno['center'][0],
            'image/object/center/y' : anno['center'][1],
            'image/object/scale'    : anno['scale'],
        }

    def image_to_bytes(self, image):
        buf = io.BytesIO()
        image.save(buf, format="JPEG")
        return buf.getvalue()

    def crop_roi(self, image, features, margin=0.2):
        img_w, img_h = image.size
        kp_x  = torch.tensor(features['image/object/parts/x'], dtype=torch.int32)
        kp_y  = torch.tensor(features['image/object/parts/y'], dtype=torch.int32)
        body_h = features['image/object/scale'] * 200.0
        mx, my = kp_x[kp_x>0], kp_y[kp_y>0]
        extra  = int(body_h * margin)
        exmin  = max(int(mx.min())-extra, 0)
        eymin  = max(int(my.min())-extra, 0)
        exmax  = min(int(mx.max())+extra, img_w)
        eymax  = min(int(my.max())+extra, img_h)
        cropped = image.crop((exmin, eymin, exmax, eymax))
        nw, nh  = exmax-exmin, eymax-eymin
        return cropped, (kp_x.float()-exmin)/nw, (kp_y.float()-eymin)/nh

    def generate_2d_gaussian(self, height, width, y0, x0, visibility=2, scale=12):
        sigma = self.gaussian_sigma          # ← 실험 포인트
        hm    = torch.zeros((height, width), dtype=torch.float32)
        xmin,ymin = x0-3*sigma, y0-3*sigma
        xmax,ymax = x0+3*sigma, y0+3*sigma
        if xmin>=width or ymin>=height or xmax<0 or ymax<0 or visibility==0:
            return hm
        size   = int(6*sigma+1)
        g      = torch.arange(0, size, dtype=torch.float32)
        xg, yg = torch.meshgrid(g, g, indexing='xy')
        cx, cy = size//2, size//2
        patch  = torch.exp(-(((xg-cx)**2+(yg-cy)**2)/(2*sigma**2)))*scale
        pxmin,pymin = max(0,-xmin), max(0,-ymin)
        pxmax,pymax = min(xmax,width)-xmin, min(ymax,height)-ymin
        hxmin,hymin = max(0,xmin), max(0,ymin)
        hxmax,hymax = min(xmax,width), min(ymax,height)
        hm[hymin:hymax, hxmin:hxmax] = patch[int(pymin):int(pymax), int(pxmin):int(pxmax)]
        return hm

    def make_heatmaps(self, features, kp_x, kp_y, heatmap_shape):
        v = torch.tensor(features['image/object/parts/v'], dtype=torch.float32)
        x = torch.round(kp_x*heatmap_shape[1]).to(torch.int32)
        y = torch.round(kp_y*heatmap_shape[0]).to(torch.int32)
        hms = [self.generate_2d_gaussian(
                   heatmap_shape[0], heatmap_shape[1],
                   int(y[i].item()), int(x[i].item()), int(v[i].item()))
               for i in range(heatmap_shape[2])]
        return torch.stack(hms, dim=0)

print('슝=3')
```

## 5. Dataset & DataLoader

```python
def parse_one_annotation(anno, image_dir):
    return {
        'filename'         : anno['image'],
        'filepath'         : os.path.join(image_dir, anno['image']),
        'joints_visibility': anno['joints_vis'],
        'joints'           : anno['joints'],
        'center'           : anno['center'],
        'scale'            : anno['scale'],
    }

class MPIIDataset(Dataset):
    def __init__(self, annotation_file, image_dir, transform=None):
        self.transform = transform
        with open(annotation_file) as f:
            annos = json.load(f)
        self.annotations = [parse_one_annotation(a, image_dir) for a in annos]
    def __len__(self): return len(self.annotations)
    def __getitem__(self, idx):
        anno  = self.annotations[idx]
        image = Image.open(anno['filepath']).convert('RGB')
        if self.transform:
            return self.transform({'image': image, 'annotation': anno})
        return image, anno

def create_dataloader(annotation_file, image_dir, batch_size,
                      num_heatmap, is_train, gaussian_sigma):
    set_seed()   # ★ DataLoader 생성 전 시드 재고정
    preprocess = Preprocessor(
        image_shape=IMAGE_SHAPE,
        heatmap_shape=(HEATMAP_SIZE[0], HEATMAP_SIZE[1], num_heatmap),
        is_train=is_train,
        gaussian_sigma=gaussian_sigma,
    )
    ds = MPIIDataset(annotation_file, image_dir, transform=preprocess)
    return DataLoader(
        ds,
        batch_size=batch_size,
        shuffle=is_train,
        num_workers=4,
        pin_memory=True,
        drop_last=False,
        prefetch_factor=2,
        worker_init_fn=seed_worker,
        generator=make_generator(),
    )

print('슝=3')
```

## 6. SimpleBaseline 모델

```python
class SimpleBaseline(nn.Module):
    def __init__(self, num_heatmap=16, pretrained=True):
        super().__init__()
        resnet = tv_models.resnet50(
            weights=tv_models.ResNet50_Weights.DEFAULT if pretrained else None
        )
        self.backbone = nn.Sequential(
            resnet.conv1, resnet.bn1, resnet.relu, resnet.maxpool,
            resnet.layer1, resnet.layer2, resnet.layer3, resnet.layer4,
        )
        def _deconv(in_ch, out_ch):
            return nn.Sequential(
                nn.ConvTranspose2d(in_ch, out_ch, 4, 2, 1, bias=False),
                nn.BatchNorm2d(out_ch, momentum=0.1),
                nn.ReLU(inplace=True),
            )
        self.deconv = nn.Sequential(
            _deconv(2048, 256), _deconv(256, 256), _deconv(256, 256),
        )
        self.final = nn.Conv2d(256, num_heatmap, 1)
        for m in self.deconv.modules():
            if isinstance(m, nn.ConvTranspose2d):
                nn.init.normal_(m.weight, std=0.001)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias,   0)
        nn.init.normal_(self.final.weight, std=0.001)
        nn.init.constant_(self.final.bias, 0)

    def forward(self, x):
        return [self.final(self.deconv(self.backbone(x)))]

# 동작 확인
_d = torch.zeros(2, 3, 256, 256)
_m = SimpleBaseline(pretrained=False)
print(f"출력 shape: {_m(_d)[0].shape}")  # (2, 16, 64, 64)
print('슝=3')
```

## 7. Trainer

```python
class Trainer:
    def __init__(self, model, epochs, batch_size, lr):
        self.model      = model
        self.epochs     = epochs
        self.optimizer  = torch.optim.Adam(model.parameters(), lr=lr)
        self.current_lr = lr
        self.lowest_val = math.inf
        self.last_val   = math.inf
        self.patience   = 0
        self.best_path  = None

    def _loss(self, labels, outputs):
        loss = 0
        for out in outputs:
            w    = (labels > 0).float() * POS_WEIGHT + 1
            loss += ((labels - out)**2 * w).mean() / labels.shape[0]
        return loss

    def _lr_decay(self):
        if self.patience >= 10:
            self.current_lr /= 10.0
            self.patience    = 0
        elif self.last_val == self.lowest_val:
            self.patience    = 0
        self.patience += 1
        for pg in self.optimizer.param_groups:
            pg['lr'] = self.current_lr

    def run(self, train_loader, val_loader, device, exp_name):
        self.model.to(device)
        tr_losses, vl_losses = [], []
        for ep in range(1, self.epochs+1):
            self._lr_decay()
            print(f"  [Epoch {ep}/{self.epochs}] lr={self.current_lr:.2e}")
            # train
            self.model.train()
            tot, n = 0.0, 0
            for imgs, lbls in train_loader:
                imgs, lbls = imgs.to(device), lbls.to(device)
                self.optimizer.zero_grad()
                loss = self._loss(lbls, self.model(imgs))
                loss.backward(); self.optimizer.step()
                tot += loss.item(); n += 1
                if n % 100 == 0:
                    print(f"    batch {n:4d}  loss {loss.item():.4f}  avg {tot/n:.4f}")
            tr = tot/n; tr_losses.append(tr)
            # val
            self.model.eval()
            tot, n = 0.0, 0
            with torch.no_grad():
                for imgs, lbls in val_loader:
                    imgs, lbls = imgs.to(device), lbls.to(device)
                    l = self._loss(lbls, self.model(imgs)).item()
                    if not math.isnan(l):
                        tot += l; n += 1
            vl = tot/n if n else float('nan'); vl_losses.append(vl)
            print(f"    → train {tr:.4f}  val {vl:.4f}")
            if vl < self.lowest_val:
                self.lowest_val = vl
                path = os.path.join(MODEL_PATH, f'{exp_name}_ep{ep}_val{vl:.4f}.pt')
                torch.save(self.model.state_dict(), path)
                self.best_path = path
                print(f"    ✓ saved: {path}")
            self.last_val = vl
        return self.best_path, tr_losses, vl_losses

print('슝=3')
```

## 8. Gaussian Sigma Ablation 실험

> sigma 값만 바꾸고 나머지는 모두 동일 조건으로 실험합니다.  
> 실험마다 `set_seed()` 를 재호출해 초기 가중치까지 동일하게 맞춥니다.


```python
sigma_results = {}  # { 'SIGMA_A': (best_path, tr_losses, vl_losses) }

for cond_name, sigma_val in SIGMA_CONDITIONS.items():
    print(f"\n{'='*55}")
    print(f"  실험: {cond_name}  |  sigma = {sigma_val}")
    print(f"{'='*55}")

    set_seed()   # ★ DataLoader 시드 고정
    train_loader = create_dataloader(TRAIN_JSON, IMAGE_PATH, BATCH_SIZE,
                                     NUM_HEATMAP, is_train=True,  gaussian_sigma=sigma_val)
    val_loader   = create_dataloader(VALID_JSON, IMAGE_PATH, BATCH_SIZE,
                                     NUM_HEATMAP, is_train=False, gaussian_sigma=sigma_val)

    set_seed()   # ★ 모델 초기 가중치 고정
    model   = SimpleBaseline(num_heatmap=NUM_HEATMAP, pretrained=True)
    trainer = Trainer(model, EPOCHS, BATCH_SIZE, lr=LR)
    best, tr, vl = trainer.run(train_loader, val_loader, device,
                               exp_name=f'sigma_{cond_name}')
    sigma_results[cond_name] = (best, tr, vl)
    print(f"  완료 | best val = {min(vl):.4f}")

print("\n[Sigma Ablation 전체 완료]")
```

## 9. Loss 비교 그래프

```python
ep_range = range(1, EPOCHS+1)
colors   = {'SIGMA_A':'#e74c3c', 'SIGMA_B':'#2ecc71', 'SIGMA_C':'#3498db'}
labels   = {'SIGMA_A':'sigma=1', 'SIGMA_B':'sigma=2 (baseline)', 'SIGMA_C':'sigma=3'}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for cond, (_, tr, vl) in sigma_results.items():
    axes[0].plot(ep_range, tr, 'o-',  color=colors[cond], label=labels[cond], lw=2)
    axes[1].plot(ep_range, vl, 's--', color=colors[cond], label=labels[cond], lw=2)

for ax, title in zip(axes, ['Train Loss', 'Validation Loss']):
    ax.set_title(f'Sigma Ablation — {title}', fontsize=13)
    ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
    ax.legend(); ax.grid(True, alpha=0.4)

plt.suptitle('Gaussian Sigma Ablation Study', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(MODEL_PATH, 'ablation_sigma_loss.png'), dpi=150)
plt.show()

# 수치 표
print(f"\n{'':>6} | {'sigma=1 (tr/vl)':>22} | {'sigma=2 (tr/vl)':>22} | {'sigma=3 (tr/vl)':>22}")
print('-'*80)
for ep in range(EPOCHS):
    print(f"ep {ep+1:>2}", end='')
    for cond in SIGMA_CONDITIONS:
        _, tr, vl = sigma_results[cond]
        print(f" | {tr[ep]:>9.4f}/{vl[ep]:>9.4f}   ", end='')
    print()
print()
print("Best val loss:")
for cond in SIGMA_CONDITIONS:
    _, _, vl = sigma_results[cond]
    print(f"  {labels[cond]:>25} : {min(vl):.4f}")
```

## 10. 추론 & 시각화 유틸

```python
MPII_BONES = [
    [0,1],[1,2],[2,6],[3,6],[3,4],[4,5],
    [6,7],[7,8],[8,9],
    [10,11],[11,12],[7,12],[7,13],[13,14],[14,15]
]

def extract_keypoints(heatmaps):
    H, W, C = heatmaps.shape
    flat    = heatmaps.reshape(-1, C)
    idx     = torch.argmax(flat, dim=0)
    y       = idx // W;  x = idx - W * y
    kps     = torch.stack([x, y], dim=1).float()
    padded  = F.pad(heatmaps.permute(2,0,1),(1,1,1,1)).permute(1,2,0)
    adj = []
    for i, kp in enumerate(kps):
        mx, my = int(kp[0])+1, int(kp[1])+1
        patch  = padded[my-1:my+2, mx-1:mx+2, i].clone()
        patch[1,1] = 0
        pidx   = torch.argmax(patch.reshape(-1)).item()
        ny, nx = divmod(pidx, 3)
        adj.append((kp[0].item()+(nx-1)/4.0, kp[1].item()+(ny-1)/4.0))
    return torch.clamp(torch.tensor(adj), 0, H) / H

def predict(model, image_path, device):
    image = Image.open(image_path).convert('RGB')
    tf    = transforms.Compose([
        transforms.Resize((256,256)),
        transforms.ToTensor(),
        transforms.Lambda(lambda x: x*2-1),
    ])
    inp = tf(image).unsqueeze(0).to(device)
    model.eval()
    with torch.no_grad():
        outs = model(inp)
    hm = outs[-1].squeeze(0).permute(1,2,0).cpu()
    return image, extract_keypoints(hm)

def draw_skeleton(ax, image, keypoints, title=''):
    img_np = np.array(image)
    joints = [(kp[0]*img_np.shape[1], kp[1]*img_np.shape[0]) for kp in keypoints]
    ax.imshow(img_np)
    for k, (i,j) in enumerate(MPII_BONES):
        ax.plot([joints[i][0], joints[j][0]],
                [joints[i][1], joints[j][1]],
                lw=3, alpha=0.8, color=plt.cm.tab20.colors[k % 20])
    for jx, jy in joints:
        ax.scatter(jx, jy, s=30, c='white', edgecolors='black', zorder=5)
    ax.set_title(title, fontsize=11, fontweight='bold')
    ax.axis('off')

print('슝=3')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, (cond, (best_path, _, vl)) in zip(axes, sigma_results.items()):
    m = SimpleBaseline(num_heatmap=NUM_HEATMAP, pretrained=False)
    m.load_state_dict(torch.load(best_path, map_location=device))
    m.to(device)
    img, kp = predict(m, test_image_path, device)
    draw_skeleton(ax, img, kp,
                  title=f"{labels[cond]}\nbest val={min(vl):.4f}")

plt.suptitle('Gaussian Sigma Ablation — Pose Estimation 비교',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(MODEL_PATH, 'ablation_sigma_pose.png'), dpi=150)
plt.show()
```

Gaussian Sigma의 역할

## Gaussian Sigma의 역할
### sigma가 작을수록 heatmap이 뾰족해 관절 위치를 정밀하게 표현하지만, 학습 신호가 약해질 수 있다고 예상했다. 그러나 실험 결과 pretrained backbone의 강력한 feature extraction 능력 덕분에 sigma=1에서도 안정적인 학습이 가능했다. sigma가 클수록 heatmap이 넓어져 인접 관절 간 heatmap이 겹치고, 모델이 관절 위치를 특정하기 어려워진다는 것을 수치로 확인했다.

## Loss 스케일 주의
### sigma가 커지면 heatmap 값의 분포가 넓어져 loss 절대값 자체가 커진다. 따라서 조건 간 loss를 단순 비교할 때는 이 점을 감안해야 한다.
